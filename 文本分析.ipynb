# 浪姐官微指标分析

```python
df = pd.read_csv('drive/My Drive/weibo/浪姐官方微博.csv')
```

```python
df.columns
```

```python
# 列的名字移位，重新命名
df.columns = ['id', 'bid', '正文', '头条文章url', '原始图片url', '视频url', '位置',
       '日期', '工具', '点赞数', '评论数', '转发数', '话题', '@用户', '时间']
```

```python
# 将日期列转化成日期格式
df['日期'] = pd.to_datetime(df['日期'])
```

```python
# 排序
df_sorted = df.sort_values(by='日期')
```

```python
# 选择符合时间范围的数据
df1 = df_sorted.loc[(df['日期'] >= '2023-05-05') & (df['日期'] <= '2023-07-22')]
```

```python
# 删除最后一行重复数据
df1 = df1.drop(df1.columns[-1], axis=1)
```

```python
df1.head()
```

```python
# 事件期间点赞、评论、转发数据
print('点赞总数:',df1.点赞数.sum(),' 日平均数：',df1.点赞数.sum()/78)
print('评论总数:',df1.评论数.sum(),' 日平均数：',df1.评论数.sum()/78)
print('转发总数:',df1.转发数.sum(),' 日平均数：',df1.转发数.sum()/78)
```

# 其他相关微博指标分析

```python
df_other = pd.read_csv('drive/My Drive/weibo/含关键词微博.csv')
```

```python
df_other.head()
```

```python
df_other.columns
```

```python
# 删除第一行数据
df_other = df_other.drop(df_other.columns[0], axis=1)
```

```python
# 将日期列转化成日期格式
df_other['发布时间'] = pd.to_datetime(df_other['发布时间'])
```

```python
# 排序
df_other_sorted = df_other.sort_values(by='发布时间')

# 重设index
df_other_sorted = df_other_sorted.reset_index(drop=True)
```

```python
# 在数据收集的过程当中混杂了一些超出浪姐开播头两天的不完整数据，故清除
df_other_sorted = df_other_sorted.loc[(df_other_sorted['发布时间'] <= '2023-05-07')]
```

```python
# 事件期间点赞、评论、转发数据
print('点赞总数:',df_other_sorted.点赞数.sum(),' 日平均数：',df_other_sorted.点赞数.sum()/2)
print('评论总数:',df_other_sorted.评论数.sum(),' 日平均数：',df_other_sorted.评论数.sum()/2)
print('转发总数:',df_other_sorted.转发数.sum(),' 日平均数：',df_other_sorted.转发数.sum()/2)
```

```python
# 检查是否包含了官微的数据
df_other_sorted[df_other_sorted.user_id == '7397600914']
```

```python
print('帖子总数：', len(df_other_sorted))
print('帖子日平均数：', len(df_other_sorted)/2)
```

```python
print('热门帖子数量：',len(df_other_sorted[df_other_sorted.点赞数 >= 1000]))
print(f"热门帖子占比：{len(df_other_sorted[df_other_sorted.点赞数 >= 1000])/len(df_other)*100}%")
```

# 情感分析

```python
# 清除 #文本# 模式的内容
df_other_sorted['微博正文'] = df_other_sorted['微博正文'].str.replace(r'#.*?#', '')

# 清除广告
df_other_final = df_other_sorted[df_other_sorted.微博正文.str.contains('芒果tv会员年卡') == False]
```

```python

from transformers import pipeline

pipe = pipeline("text-classification", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",max_length=512, truncation=True)
#pipe = pipeline('zero-shot-classification', model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")
```

```python
from tqdm import tqdm
sentiment_list = []
```

```python
for tweet in tqdm(df_other_final['微博正文']):
   #try:
     sentiment = pipe(tweet)[0]['label']
     sentiment_list.append(sentiment)
   #except:
     #pass
```

```python
df_other_final['sentiment'] = sentiment_list
```

```python
# 情感标签分布
df_other_final['sentiment'].value_counts()
```
